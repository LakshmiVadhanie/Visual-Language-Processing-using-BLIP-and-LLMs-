```python
import os
import tarfile

# URLs
images_url = "https://openi.nlm.nih.gov/imgs/collections/NLMCXR_png.tgz"
reports_url = "https://openi.nlm.nih.gov/imgs/collections/NLMCXR_reports.tgz"

# Paths
data_path = "./data"
os.makedirs(data_path, exist_ok=True)

# Function to download a file
def download_file(url, save_path):
    file_name = url.split("/")[-1]
    file_path = os.path.join(save_path, file_name)
    if not os.path.exists(file_path):
        os.system(f"wget -P {save_path} {url}")
        print(f"Downloaded {file_name}")
    else:
        print(f"{file_name} already exists.")
    return file_path

# Function to list contents of .tgz
def list_tgz_contents(tgz_path):
    with tarfile.open(tgz_path, "r:gz") as tar:
        files = tar.getnames()
        print(f"Contents of {tgz_path}:")
        for file in files[:10]:  # Display a few files for clarity
            print(file)
        return files

# Function to extract specific files
def extract_specific_files(tgz_path, extract_to, files_to_extract):
    with tarfile.open(tgz_path, "r:gz") as tar:
        for file in files_to_extract:
            try:
                tar.extract(file, path=extract_to)
                print(f"Extracted {file} to {extract_to}")
            except KeyError:
                print(f"{file} not found in the archive.")

# Download archives
images_tgz = download_file(images_url, data_path)
reports_tgz = download_file(reports_url, data_path)

# List contents of archives
image_files = list_tgz_contents(images_tgz)
report_files = list_tgz_contents(reports_tgz)

selected_image_files = image_files
selected_report_files = report_files

# Extract selected files
images_path = os.path.join(data_path, "images_subset")
reports_path = os.path.join(data_path, "reports_subset")
os.makedirs(images_path, exist_ok=True)
os.makedirs(reports_path, exist_ok=True)

extract_specific_files(images_tgz, images_path, selected_image_files)
extract_specific_files(reports_tgz, reports_path, selected_report_files)
```
